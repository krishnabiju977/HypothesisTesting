```python
#Ho:Significant difference in the diameter of the cutlet
# H1: No significant difference in the diameter of the cutlet
#Test statistic t=(x1-x2)/sqrt(s1^2/n1+s2^2/n2)
# Where:
 #x1=mean of unit A
 #x2=mean of unit B
 #S1^2=sample variance of unit A
 #S2^2= sample variance of unit B
 #n1=size of unit A
 #n2=size of unit B
```

```python
unitA=pd.Series(cutlets.iloc[:,0])
unitA
```

```python
unitB=pd.Series(cutlets.iloc[:,1])
unitB
```

```python
p_value=stats.ttest_ind(unitA,unitB)
p_value
```

```python
#So we get the t value as 0.722 and p value as0.47223.
#From t table we get the critical value  for 0.05 significance level and degrees of freedom 34 as 2.402
#Here critical value is greater than test statistic t value.so fail to reject the null hypothesis.
#Accept Ho ,i.e there is a significant difference the diameter  of the cutlet.
```
